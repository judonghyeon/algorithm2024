### 1) 문제 정의 (MD)

문제 이름 : 이진 탐색(순환구조)

문제 설명 : 정렬된 리스트와 키값, 탐색 범위가 제공되었을 때 리스트에 키값을 반환하는 알고리즘을 작성

문제 예시 : list = [1, 3, 8, 13, 13, 16, 21, 26, 27, 30, 33, 36, 39, 41, 44, 49] 와 키값 33, 범위 0부터 len(list)-1 까지 주어진 경우 리턴값은 10 (인덱스)

### 2) 알고리즘 설명  (MD)

범위를 반으로 줄여 키값이 어느 위치 (반으로 나눈 기준보다 작은지, 큰지, 같은지)에 있는지 판단 후 경우에 따라 함수를 순환적으로 호출하여 키값이 몇 번째 인덱스에 있는지 반환함.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : binary_search

매개변수 : 정렬된 리스트 A, 키 값 key, 범위 low, high

(예시) : binary_search(listA, 33, 0, len(listA)-1)

출력 : 리스트에 있는 키 값의 인덱스 (없으면 -1)

수행 순서 : 
1. (탐색할 범위가 남아있는 동안 : 범위의 최대와 최소를 더해 2로 나눈 값(mid)을 키 값과 비교한다.
2. 키 값과 같다면 인덱스를 반환하고, 키 값이 더 크거나 작다면 함수의 매개변수중 범위값을 달리하여 함수를 다시 호출한다.)
3. 범위를 모두 탐색했음에도 키 값을 발견하지 못했다면 -1을 반환한다.

### 5) 알고리즘 코드 (PY,C)

```python
def binary_search(A, key, low, high) :
    if(low <= high) :
        mid = (low + high) // 2
        if key == A[mid] :
            return mid
        elif key < A[mid] :
            return binary_search(A, key, low, mid-1)
        else :
            return binary_search(A, key, mid+1, high)
    return -1
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
listA = [1,2,3,4,5,6,7]
print("4의 인덱스 : ", binary_search(listA, 4, 0, len(listA)-1))
```
