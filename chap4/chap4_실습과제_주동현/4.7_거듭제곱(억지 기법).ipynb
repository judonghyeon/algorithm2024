### 1) 문제 정의 (MD)

문제 이름 : 거듭제곱(억지 기법)

문제 설명 : 밑 X과 지수 n을 매개변수로 하여 x의 n제곱을 반환하는 함수

문제 예시 : x = 5, n = 2라면 5^2 = 25가 반환되어야 함.

### 2) 알고리즘 설명  (MD)

반복문을 통해 x를 n번 곱하는 연산을 수행함.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : slow_power(x, n)

매개변수 : 실수 x, n

(예시) : slow_power(5, 2)

출력 : x의 n제곱

수행 순서 : 
1. 결과를 저장하는 변수를 생성한다.
2. 반복문을 통해 결과에 x를 n번 곱한 값을 반환한다.

### 5) 알고리즘 코드 (PY,C)

```python
def slow_power(x, n) :
    result = 1.0
    for i in range(n) :
        result *= x
    return result
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
print(slow_power(10,8))
```
