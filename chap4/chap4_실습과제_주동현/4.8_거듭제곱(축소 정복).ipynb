### 1) 문제 정의 (MD)

문제 이름 : 거듭제곱(축소 정복 기법)

문제 설명 : 밑 X과 지수 n을 매개변수로 하여 x의 n제곱을 반환하는 함수

문제 예시 : x = 5, n = 2라면 5^2 = 25가 반환되어야 함.

### 2) 알고리즘 설명  (MD)

곱셈 횟수 n에 따라 매개변수를 다르게 하여 함수를 순환적으로 호출하여 x의 n제곱을 반환함.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : power(x, n)

매개변수 : 실수 x, n

(예시) : power(5, 2)

출력 : x의 n제곱

수행 순서 : 
1. n이 0이라면 1을 반환한다.
2. n이 짝수라면 x*x, n//2로 매개변수를 변경하여 함수를 다시 호출한다.
3. n이 홀수하면 x*x, (n-1)//2로 매개변수를 변경하여 함수를 다시 호출한다.

### 5) 알고리즘 코드 (PY,C)

```python
def power(x, n) :
   if n == 0 :
        return 1
   elif (n%2) == 0 :
        return power(x*x, n//2)
   else :
        return x*power(x*x, (n-1)//2)
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
print(power(2,8))
```
