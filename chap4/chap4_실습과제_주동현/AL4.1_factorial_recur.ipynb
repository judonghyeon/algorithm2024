사용자로부터 n의 값을 입력 받고 n의 값의 -1한 값의 팩토리얼을 반복하여 1일 때까지 구한다. 

##### 3) 손으로 푼 예제


### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 명 : factorial_recur(n):

매개변수 : 팩토리얼 n


출력 : 팩토리얼 n의 값

수행 순서 :

1. n의 값이 1일 경우 return 1 
2. 아닐 경우 n* factorial_recur(n-1)을 하여 n1일 때까지 반복 계산

#### 5) 알고리즘 코드 (PY,C)

```python
def factorial_recur(n):
    if n == 1:
        return 1
    else:
        return n * factorial_recur(n - 1)
```

####   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def factorial_recur(n):
    if n == 1:
        return 1
    else:
        return n * factorial_recur(n - 1)

num = int(input("팩토리얼을 계산할 숫자: "))

result = factorial_recur(num)
print("팩토리얼 값 :",result)
```
