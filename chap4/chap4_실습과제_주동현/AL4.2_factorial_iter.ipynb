### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 명 : factorial_recur(n):

매개변수 : 팩토리얼 n


출력 : 팩토리얼 n의 값

수행 순서 :

1. n의 값 만큼 반복하여 1(result)*반복횟수(1) > 1(result)* (반복횟수)2 > 2(result)*(반복횟수)3...계산

### 5) 알고리즘 코드 (PY,C)

```python
def factorial_iter(n):
    result =1
    for k in range(1,n+1):
        result = result*k
    return result
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def factorial_iter(n):
    result =1
    for k in range(1,n+1):
        result = result*k
    return result
num=int(input("팩토리얼을 계산할 숫자:"))
fac=(factorial_iter(num))
print("팩토리얼 값:",fac)
```
