1) 문제 정의 (MD)
각각 다른 순서로 노드를 방문하며, 트리의 구조에 따라 사용됩니다.
전위 순회 (Preorder Traversal): 루트 노드를 먼저 방문한 후, 왼쪽 서브트리를 전위 순회하고, 오른쪽 서브트리를 전위 순회합니다.
중위 순회 (Inorder Traversal): 왼쪽 서브트리를 중위 순회한 후에 루트 노드를 방문하고, 오른쪽 서브트리를 중위 순회합니다.
후위 순회 (Postorder Traversal): 왼쪽 서브트리를 후위 순회한 후에 오른쪽 서브트리를 후위 순회하고 마지막으로 루트 노드를 방문합니다.
이 문제는 주어진 이진트리에 대해 위의 세 가지 순회 방법 중 하나를 선택하여 노드를 순회하는 알고리즘을 작성하는 것을 목표로 합니다.

2) 알고리즘 설명  (MD)
1. 전위 순회
루트 노드를 먼저 방문합니다.
왼쪽 서브트리를 전위 순회합니다.
오른쪽 서브트리를 전위 순회합니다.
즉, "루트-왼쪽-오른쪽" 순서로 노드를 방문합니다.

2. 중위 순회
왼쪽 서브트리를 중위 순회합니다.
루트 노드를 방문합니다.
오른쪽 서브트리를 중위 순회합니다.
즉, "왼쪽-루트-오른쪽" 순서로 노드를 방문합니다.

3. 후위 순회
왼쪽 서브트리를 후위 순회합니다.
오른쪽 서브트리를 후위 순회합니다.
루트 노드를 방문합니다.
즉, "왼쪽-오른쪽-루트" 순서로 노드를 방문합니다.

5) 코드 (PY,C)


```python
#전위 순회

def preorder(n) :
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)

#중위 순회

def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)

#후위 순회

def postorder(n) :
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
```

6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# 전위 순회 함수
def preorder(n):
    if n is not None:
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)

# 중위 순회 함수
def inorder(n):
    if n is not None:
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)

# 후위 순회 함수
def postorder(n):
    if n is not None:
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')

# 테스트 코드
if __name__ == "__main__":
    # 트리 구성
    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(3)
    root.left.left = TreeNode(4)
    root.left.right = TreeNode(5)

    # 전위 순회 테스트
    print("전위 순회 결과:")
    preorder(root)
    print()

    # 중위 순회 테스트
    print("중위 순회 결과:")
    inorder(root)
    print()

    # 후위 순회 테스트
    print("후위 순회 결과:")
    postorder(root)
    print()
```

전위 순회 
모든 노드를 한 번씩 방문하므로 시간 복잡도는 O(n) 입니다. 여기서 n은 이진 트리의 노드 수입니다.

중위 순회 
중위 순회 역시 모든 노드를 한 번씩 방문하므로 시간 복잡도는 O(n) 입니다.

후위 순회 
후위 순회도 모든 노드를 한 번씩 방문하므로 시간 복잡도는 O(n) 입니다.
