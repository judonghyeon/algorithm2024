### 1) 문제 정의 (MD)

문제 이름 : 선형 조사법의 삽입 알고리즘 구현

문제 설명 : 테이블에 레코드를 저장해야함.

문제 예시 : {킷값 : 해시 주소} {{45 : 6}, {27, 1}, {88, 10}, {9, 9}, {71, 6}, {60, 8}, {46, 7}, {38, 12}, {24, 11}}일 때 45 삽입 후 71 삽입시 충돌이 발생하여 41 다음 칸에 저장되어야 함.

### 2) 알고리즘 설명  (MD)

킷값을 주소로 변환하고, 테이블의 그 주소가 비어있다면 저장하고, 아니라면 빈 곳을 찾아 저장한다.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)


### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : lp_insert(key)

매개변수 : 킷값 key

출력 : 출력 없이 테이블에 킷값을 저장함.

수행 순서 : 
1. 킷값을 해시 함수를 통해 해시 주소를 구한다.
2. 주소에 해당하는 테이블이 비어있지 않다면 다음위치를 계속 검사한다.
3. 빈 테이블을 찾으면 킷값을 저장한다.

### 5) 알고리즘 코드 (PY,C)

```python
M = 13
table = [None] * M
def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count > 0 and (table[id] != None) :
        id = (id + 1 + M) % M
        count -= 1
    if count > 0 :
        table[id] = key
    return
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
print("최초 :", table)
lp_insert(45)
print("45 삽입 :", table)
lp_insert(27)
print("27 삽입 :", table)
lp_insert(88)
print("88 삽입 :", table)
lp_insert(9)
print("9 삽입 :", table)
lp_insert(71)
print("71 삽입 :", table)
lp_insert(60)
print("60 삽입 :", table)
lp_insert(46)
print("46 삽입 :", table)
lp_insert(38)
print("38 삽입 :", table)
lp_insert(24)
print("24 삽입 :", table)
```
