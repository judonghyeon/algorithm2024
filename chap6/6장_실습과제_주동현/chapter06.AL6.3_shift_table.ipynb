## 알고리즘 설명

호스풀(Horspool) 알고리즘은 문자열 검색 알고리즘 중 하나로, 텍스트와 패턴을 입력으로 받아 패턴이 텍스트 내에서 어디에 있는지를 찾습니다. 이 알고리즘의 핵심은 패턴을 텍스트 상에서 한 칸씩 이동하면서 비교하는 것입니다.

호스풀 알고리즘의 동작은 다음과 같습니다:

1. **시프트 테이블 생성**: 패턴의 각 문자에 대해 마지막으로 등장하는 위치를 저장하는 시프트 테이블을 만듭니다.
2. **패턴 검색**: 텍스트를 왼쪽에서 오른쪽으로 이동하면서 패턴과 비교합니다.
    - 패턴의 마지막 문자와 텍스트를 비교합니다.
    - 만약 패턴의 마지막 문자와 텍스트가 일치하면, 패턴의 뒤에서부터 앞으로 비교해가며 모든 문자가 일치하는지 확인합니다.
    - 일치하지 않는 문자가 나타나면, 시프트 테이블을 사용하여 텍스트를 적절히 이동시킵니다.
    - 일치하지 않는 문자가 나타나지 않고 패턴의 모든 문자가 일치하면, 패턴이 발견된 위치(인덱스)를 반환합니다.
3. **패턴이 발견되지 않을 경우**: 텍스트를 끝까지 검색했는데도 패턴이 발견되지 않으면 -1을 반환합니다.

호스풀 알고리즘은 일반적으로 단순한 구현과 빠른 속도로 알려져 있습니다. 그러나 최악의 경우에는 O(mn)의 시간 복잡도를 가지기 때문에 패턴과 텍스트의 길이가 매우 큰 경우에는 다른 알고리즘을 고려해야 할 수 있습니다.

## 복잡도 분석

### 호스풀 알고리즘의 복잡도

호스풀 알고리즘의 시간 복잡도는 입력 텍스트와 패턴의 길이에 따라 다릅니다.

- **최선, 평균 시간 복잡도: O(n/m)**, 여기서 n은 텍스트의 길이, m은 패턴의 길이입니다.
- **최악 시간 복잡도: O(mn)**, 텍스트의 모든 문자가 패턴과 일치하는 경우입니다.
- **공간 복잡도: O(1)**, 추가적인 공간이 필요하지 않습니다.

### 복잡도 분석

- 호스풀 알고리즘은 각 단계마다 패턴을 텍스트에 이동시키며 비교합니다. 
- 최악의 경우, 각 단계에서 패턴이 텍스트와 한 번에 한 칸씩 이동하므로 O(mn)의 시간 복잡도를 가집니다.
- 그러나 일반적으로 패턴이 텍스트와 매우 다른 경우가 많기 때문에 실제로는 O(n/m)의 시간 복잡도를 가집니다.

호스풀 알고리즘은 단순한 구현과 높은 효율성을 가지고 있지만, 최악의 경우에는 성능이 떨어질 수 있습니다.
